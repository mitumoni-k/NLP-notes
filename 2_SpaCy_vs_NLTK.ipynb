## Sentence / Word Tokenization using SpaCy

```python
!pip install spacy
!python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Hi, My name is Mitumoni Kalita. I am currently a final year CSE UG student at Jorhat Engineering College. I love to code, cook and dance...")

## SECTENCE TOKENIZATION - converting this paragraph into invidual sentences
## we can see here that we considered an object as 'doc' which has a property called 'sents', hence, spacy is a OOP library

print('SENTENCES :')
for sentence in doc.sents:
    print(sentence)

## WORD TOKENIZATION - converting each sentences into individual words



print(' ')    
print('WORDS :')
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

## Sentence / Word Tokenization using NLTK library

```python
import nltk
nltk.download('punkt')
```

```python
## converting paragraph into individual sentences - SENTENCE TOKENIZATION

from nltk.tokenize import sent_tokenize

sent_tokenize("Hi, My name is Mitumoni Kalita. I am currently a final year CSE UG student at Jorhat Engineering College. I love to code, cook and dance...")
```

```python
## converting the paragraph into an array of individual words - WORD TOKENIZATION

from nltk.tokenize import word_tokenize

word_tokenize("Hi, My name is Mitumoni Kalita. I am currently a final year CSE UG student at Jorhat Engineering College. I love to code, cook and dance...")
```
